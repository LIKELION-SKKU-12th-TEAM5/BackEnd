```python
import os

from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
from langchain_community.document_loaders import TextLoader
from langchain_community.document_loaders import DirectoryLoader
from collections import Counter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

os.environ["OPENAI_API_KEY"] = APIKEY
```

```python


loader = DirectoryLoader('/content/data_1', glob="*.txt", loader_cls=TextLoader)
loader1 = DirectoryLoader('/content/drive/MyDrive/주거', glob="*.txt", loader_cls=TextLoader)
loader2 = DirectoryLoader('/content/drive/MyDrive/진로', glob="*.txt", loader_cls=TextLoader)

documents = loader.load()
documents1 = loader1.load()
documents2 = loader2.load()

print("문서의 개수 :", len(documents))
print("문서의 개수 :", len(documents1))

print("document :", documents[1])
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
texts = text_splitter.split_documents(documents)
texts1 = text_splitter.split_documents(documents1)
texts2 = text_splitter.split_documents(documents2)

print('분할된 텍스트 개수 : ', len(texts))
print('분할된 텍스트 개수 : ', len(texts1))
```

```python
# SentenceTransformer를 직접 사용하는 대신 HuggingFaceEmbeddings로 감싸줍니다.
embedding_function = HuggingFaceEmbeddings(
    model_name='xlm-r-bert-base-nli-stsb-mean-tokens'
)

# 첫 번째 Chroma 데이터베이스 생성
vectordb = Chroma.from_documents(
    documents=texts,
    embedding=embedding_function,  # 감싼 임베딩 함수를 사용합니다.
    persist_directory="/content/drive/MyDrive/Colab Notebooks/db1"  # 첫 번째 데이터베이스의 데이터를 저장할 디렉토리
)

# 두 번째 Chroma 데이터베이스 생성
vectordb1 = Chroma.from_documents(
    documents=texts1,
    embedding=embedding_function,  # 감싼 임베딩 함수를 사용합니다.
    persist_directory="/content/drive/MyDrive/Colab Notebooks/db2"  # 두 번째 데이터베이스의 데이터를 저장할 디렉토리
)

vectordb2 = Chroma.from_documents(
    documents=texts2,
    embedding=embedding_function,  # 감싼 임베딩 함수를 사용합니다.
    persist_directory="/content/drive/MyDrive/Colab Notebooks/db3"  # 두 번째 데이터베이스의 데이터를 저장할 디렉토리
)


# 각 데이터베이스에 대해 retriever 생성
retriever = vectordb.as_retriever()
retriever1 = vectordb1.as_retriever()
retriever2 = vectordb1.as_retriever()
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=ChatOpenAI(model_name="gpt-3.5-turbo", temperature = 0),
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)

qa_chain1 = RetrievalQA.from_chain_type(
    llm=ChatOpenAI(model_name="gpt-3.5-turbo", temperature = 0),
    chain_type="stuff",
    retriever=retriever1,
    return_source_documents=True
)

qa_chain2 = RetrievalQA.from_chain_type(
    llm=ChatOpenAI(model_name="gpt-3.5-turbo", temperature = 0),
    chain_type="stuff",
    retriever=retriever2,
    return_source_documents=True
)

def get_chatbot_response(chatbot_response):
    result = ""
    result = result + chatbot_response['result'].strip()
    return result
```
